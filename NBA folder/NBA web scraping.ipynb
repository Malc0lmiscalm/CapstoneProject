```python
# Opening the already created 1996.html in readmode.
with open ("mvp/1996.html") as f:
    page = f.read()
```

We can inspect particular elements in our webpage by rclicking them.

```python
soup = BeautifulSoup(page, "html.parser")
soup.find('tr', class_="over_header").decompose() # What this is doing : removing the overheader in our desired tables.
```

the whole line is an html elementm, tr is the tag, which also has a class called over_header.
An id in html is a globally unique property. The table we want has id=mvp

```python
mvp_table = soup.find(id="mvp") 
```

```python
import pandas as pd
mvp_1996 = pd.read_html(str(mvp_table))[0] # for some reason we need the [0] at the end.
mvp_1996
```

```python
# Once we are happy with our test year, we can the for loop for each year in our range.
# Create a list of dataframes that in the mvp folder.
dfs = []
for year in years:
    with open("mvp/{}.html".format(year)) as f:
        page = f.read()
    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_='over_header').decompose()
    mvp_table = soup.find(id="mvp")
    mvp = pd.read_html(str(mvp_table))[0]
    mvp["Year"] = year # We need the year each entry came from. We need this step first before rearranging the columns
    dfs.append(mvp)
```

```python
mvps = pd.concat(dfs)

# Rearranging the columns so that Year is the second column
year_col = mvps['Year']
mvps = mvps.drop(columns=['Year'])
mvps.insert(loc=1, column='Year', value=year_col)

mvps.sample(10)
```

```python
# Saving our pandas dataframe back to a csv – useful to upload to Tableau, SQl etc.
mvps.to_csv("mvps.csv")
```

## Finally we need th eteam stadnings per year.

```python
# We don't mind the Javascript rendered sections and thus won't use selenium. We are scraping the Divison Standing tables.
import requests
team_stats_url = "https://www.basketball-reference.com/leagues/NBA_{}_standings.html"
for year in years:
    url = team_stats_url.format(year)

    data = requests.get(url)

    with open("team/{}.html".format(year), "w+") as f:
        f.write(data.text)
```

```python
from bs4 import BeautifulSoup
import pandas as pd
years = list(range(1991, 2023))

dfs = []
for year in years:
    with open("team/{}.html".format(year)) as f:
        page = f.read()

    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="thead").decompose()
    eastern_table = soup.find_all(id="div_divs_standings_E")[0]  # id will be found in the html elemnt starting <table class="
    eastern_df = pd.read_html(str(eastern_table))[0]
    eastern_df["Year"] = year # Disambiguate while forlooping
    eastern_df["Team"] = eastern_df["Eastern Conference"] # We are creating(poping at the rightmost column) a new column call Team which in our scraped table is called Eastern Conference.
    del eastern_df["Eastern Conference"]
    dfs.append(eastern_df)

    soup = BeautifulSoup(page, "html.parser")
    soup.find('tr', class_="thead").decompose()
    western_table = soup.find_all(id="div_divs_standings_W")[0]  # id will be found in the html elemnt starting <table class="
    western_df = pd.read_html(str(western_table))[0]
    western_df["Year"] = year # Disambiguate while forlooping
    western_df["Team"] = western_df["Western Conference"] # 
    del western_df["Western Conference"]
    dfs.append(western_df)

teams = pd.concat(dfs)
teams.sample(50)
```

```python
teams.to_csv("teams.csv")
```
