```python
import pandas as pd
pd.set_option('display.max_columns', 28)
from nba_api.stats.static import players
list_of_players = players.get_players() # This is a list of dictionaries  
list_of_players
```

```python
res = len([ player for player in list_of_players if isinstance(player, dict)])
print("The Dictionary count : " + str(res))
```

```python
# Getting player id from the player_dict
steph = [player for player in list_of_players if player['full_name'] == 'Stephen Curry'][0]
print(steph)
bron = [player for player in list_of_players if player['full_name'] == 'LeBron James'][0]
bron
```

```python
from nba_api.stats.static import teams
team_dict = teams.get_teams()

GSW = [team for team in team_dict if team['full_name'] == 'Golden State Warriors'][0]
GSW_id = GSW['id']
GSW_id
```

```python
# Let's get Lebron's data from his winning Cav's season 2015-2016
from nba_api.stats.endpoints import playergamelog
gamelog_bron = playergamelog.PlayerGameLog(player_id = bron['id'], season = '2015')
df_gamelog_bron = gamelog_bron.get_data_frames()[0]
df_gamelog_bron
```

```python
# Let's get Stephen Curry's data from ALL of his regular season games.
from nba_api.stats.library.parameters import SeasonAll
gamelog_steph_all = playergamelog.PlayerGameLog(player_id = steph["id"], season = SeasonAll.all)
df_gamelog_steph_all = gamelog_steph_all.get_data_frames()[0]
df_gamelog_steph_all
```

```python
# Getting all the Golden State Warriors since their inception.
from nba_api.stats.endpoints import leaguegamefinder
GSW_games = leaguegamefinder.LeagueGameFinder(team_id_nullable=GSW_id, season_type_nullable='Playoffs').get_data_frames()[0]
GSW_games
```

```python
# Game play by play 
from nba_api.stats.endpoints import playbyplay
pbp = playbyplay.PlayByPlay(game_id="0042100403").get_data_frames()[0]
display(pbp.loc[400:pbp.shape[0]])
```

```python
from nba_api.stats.endpoints import leaguegamefinder

def season_str(x):
    return str(x) + '-' + str(x+1)[-2:]

for season in range(2000, 2020):
    playoff_games = leaguegamefinder.LeagueGameFinder(season_nullable=season_str(season),                                  season_type_nullable='Playoffs').get_data_frames()[0]
    
    last_game = playoff_games.loc[playoff_games['GAME_ID'] == playoff_games['GAME_ID'].max()]
    
    matchups = last_game['MATCHUP'].tolist()

    finals = playoff_games.loc[playoff_games['MATCHUP'].isin(matchups)]
    
    first_date = finals['GAME_DATE'].min()
    last_date = finals['GAME_DATE'].max()
    
    print("season:", season_str(season), ":", first_date, "-", last_date)
from nba_api.stats.endpoints import leaguegamefinder

def season_str(x):
    return str(x) + '-' + str(x+1)[-2:]

for season in range(2000, 2020):
    playoff_games = leaguegamefinder.LeagueGameFinder(season_nullable=season_str(season),                                  season_type_nullable='Playoffs').get_data_frames()[0]
    
    last_game = playoff_games.loc[playoff_games['GAME_ID'] == playoff_games['GAME_ID'].max()]
    
    matchups = last_game['MATCHUP'].tolist()

    finals = playoff_games.loc[playoff_games['MATCHUP'].isin(matchups)]
    
    first_date = finals['GAME_DATE'].min()
    last_date = finals['GAME_DATE'].max()
    
    print(f"season:", season_str(season), ":", first_date, "-", last_date)
```
