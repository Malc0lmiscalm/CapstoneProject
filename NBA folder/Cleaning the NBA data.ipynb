```python
# Clean the player.csv
players = pd.read_csv("players.csv")
players
```

```python
del players["Unnamed: 0"]
del players["Rk"]
```

```python
players.head(100)
```

We want to merge the two tables, grouping on the player names(also the year ), however some Hall of Famers from the players table have an asterisk in their name.

```python
players["Player"] = players["Player"].str.replace("*", "", regex=False)
players.head(20)
```

Note that some players play in more than one team in a single yearly season.
>Note Greg Anderson in 1991

```python
players.groupby(["Player", "Year"]).get_group(("Greg Anderson", 1991 ))
```

```python
def single_team(df):
    if df.shape[0]==1:
        return df
    else:
        row = df[df["Tm"]=="TOT"]
        row["Tm"] = df.iloc[-1,:]["Tm"]
        return row

players = players.groupby(["Player", "Year"]).apply(single_team)
players
```

```python
players.index = players.index.droplevel()
players.index = players.index.droplevel()
```
